# Analysis with Pandas

# Set Up

```python
#Set up
import pandas as pd
from pathlib import Path

#Locate files
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

#Read data files and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Combine the data into single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_data_complete.info()
```

```python
cols_to_keep = ["Student ID", "school_name", "reading_score", "math_score", "budget", "type"]
complete_data_reduced = school_data_complete[cols_to_keep]
complete_data_reduced.head()
```

```python
school_data.info()
```

# District Summary

```python
#Total number of unique schools
school_count = len(school_data.value_counts("school_name"))
school_count
```

```python
#Total students
student_count = len(complete_data_reduced.value_counts("Student ID"))
student_count
```

```python
#Total budget
budget = school_data["budget"].sum()
budget
```

```python
#Average math score
avg_math_score = round(complete_data_reduced["math_score"].mean(), 2)
avg_math_score
```

```python
#Average reading score[
avg_reading_score = round(complete_data_reduced["reading_score"].mean(), 2)
avg_reading_score
```

```python
#% passing math
pass_math = complete_data_reduced[(complete_data_reduced["math_score"] >= 70)].count()["Student ID"]
passing_math_percentage = round(pass_math / student_count * 100, 2)
passing_math_percentage
```

```python
#% passing reading
pass_read = complete_data_reduced[(complete_data_reduced["reading_score"] >= 70)].count()["Student ID"]
passing_reading_percentage = round(pass_read / student_count * 100, 2)
passing_reading_percentage
```

```python
#% overall passing (percentage of students who passed math AND reading)
passing_math_reading_count = complete_data_reduced[
    (complete_data_reduced["math_score"] >= 70) & (complete_data_reduced["reading_score"] >= 70)
].count()["Student ID"]
overall_passing_rate = round(passing_math_reading_count /  student_count * 100, 2)
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools": [school_count], 
                                 "Total Students": [student_count], 
                                 "Budget": [budget], 
                                 "Average math score": [avg_math_score], 
                                 "Average reading score": [avg_reading_score], 
                                 "% of students with passing math scores": [passing_math_percentage], 
                                 "% of students with passing reading scores": [passing_reading_percentage],
                                "% of students who passed math AND reading": [overall_passing_rate]})

district_summary
```

```python
district_summary.dtypes
```

```python
#Export
district_summary.to_csv("analysis.csv",index=False)
```

# School Summary

```python
#select all of the school types
school_type = complete_data_reduced['type'].unique()

# Print or use the unique values as needed
print(school_type)
```

```python
# Calculate the total student count per school
student_count_per_school = complete_data_reduced.groupby(["school_name"])["Student ID"].count()
student_count_per_school
```

```python
# Calculate the total school budget and per capita spending per school
school_data["Budget Per Capita"] = school_data["budget"]/school_data["size"]
budget_per_capita = school_data[["school_name", "budget", 'type',"size", "Budget Per Capita"]]
budget_per_capita.head()
```

```python
# Calculate the average math score per school
avg_math_score_grouped_by_school = round(complete_data_reduced.groupby(["school_name"])["math_score"].mean(), 2)
avg_math_score_grouped_by_school
```

```python
# Calculate the average reading score per school
avg_reading_score_grouped_by_school = round(complete_data_reduced.groupby(["school_name"])["reading_score"].mean(), 2)
avg_reading_score_grouped_by_school
```

```python
# Calculate the number of students per school with math scores of 70 or higher
passing_math_scores = complete_data_reduced[complete_data_reduced["math_score"]>=70].groupby('school_name').size().reset_index(name='# of students who passed math')
passing_math_scores.head()
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
passing_reading_scores = complete_data_reduced[complete_data_reduced["reading_score"]>=70].groupby('school_name').size().reset_index(name='# of students who passed reading')
passing_reading_scores.head()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = complete_data_reduced[(complete_data_reduced["reading_score"]>=70) & (complete_data_reduced["math_score"]>=70)].groupby('school_name').size().reset_index(name='# of students who passed reading & math')
students_passing_math_and_reading.head()
```

```python
# Merge school budget data with test score data and rename columns
school_budget_scores = pd.merge(budget_per_capita, avg_reading_score_grouped_by_school, how="left", on=["school_name", "school_name"])
school_budget_scores = pd.merge(school_budget_scores, avg_math_score_grouped_by_school, how="left", on=["school_name", "school_name"])
school_budget_scores = pd.merge(school_budget_scores, passing_reading_scores, how="left", on=["school_name", "school_name"])
school_budget_scores = pd.merge(school_budget_scores, passing_math_scores, how="left", on=["school_name", "school_name"])
school_budget_scores = pd.merge(school_budget_scores, students_passing_math_and_reading, how="left", on=["school_name", "school_name"])
school_budget_scores = school_budget_scores.rename(columns={"school_name":"School Name", 
                                                            "budget":"Budget", "size":"# of Students", 
                                                            "reading_score":"Average Reading Score", 
                                                            "math_score":"Average Math Score"})

school_budget_scores.head()
```

```python
pass_reading_percentage = round(school_budget_scores["# of students who passed reading"]/school_budget_scores["# of Students"]*100, 2)
school_budget_scores["% of students who passed reading"] = pass_reading_percentage
pass_math_percentage = round(school_budget_scores["# of students who passed math"]/school_budget_scores["# of Students"]*100, 2)
school_budget_scores["% of students who passed math"] = pass_math_percentage
pass_both_percentage = round(school_budget_scores["# of students who passed reading & math"]/school_budget_scores["# of Students"]*100, 2)
school_budget_scores["% of students who passed reading & math"] = pass_both_percentage
school_budget_scores.head()
```

```python
school_budget_scores.columns
```

```python
# Create a DataFrame called `per_school_summary`
school_summary_cols_to_keep = ['School Name', 'type', '# of Students', 'Budget', 'Budget Per Capita',
       'Average Math Score','Average Reading Score', '% of students who passed math',
       '% of students who passed reading', 
       '% of students who passed reading & math']
per_school_summary = school_budget_scores[school_summary_cols_to_keep]

#Formatting

# Display the DataFrame
per_school_summary
```

```python
#Export
per_school_summary.to_csv("school_scores_budget.csv",index=False)
```

# Lowest-Performing Schools (by % Overall Passing)

# Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby(["school_name"])["math_score"].mean().map("{:,.1f}%".format)
tenth_grader_math_scores = tenth_graders.groupby(["school_name"])["math_score"].mean().map("{:,.1f}%".format)
eleventh_grader_math_scores = eleventh_graders.groupby(["school_name"])["math_score"].mean().map("{:,.1f}%".format)
twelfth_grader_math_scores = twelfth_graders.groupby(["school_name"])["math_score"].mean().map("{:,.1f}%".format)

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.merge(ninth_grade_math_scores, tenth_grader_math_scores, on=["school_name"])
math_scores_by_grade = pd.merge(math_scores_by_grade, eleventh_grader_math_scores, on=["school_name"])
math_scores_by_grade = math_scores_by_grade.rename(columns={"math_score_x": "9th Grader Math Scores", "math_score_y": "10th Grader Math Scores", "math_score": "11th Grader Math Scores"})
math_scores_by_grade = pd.merge(math_scores_by_grade, twelfth_grader_math_scores, on=["school_name"])
math_scores_by_grade = math_scores_by_grade.rename(columns={"math_score": "12th Grader Math Scores"})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

# Reading Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grader_reading_scores = ninth_graders.groupby(["school_name"])["reading_score"].mean().map("{:,.1f}%".format)
tenth_grader_reading_scores = tenth_graders.groupby(["school_name"])["reading_score"].mean().map("{:,.1f}%".format)
eleventh_grader_reading_scores = eleventh_graders.groupby(["school_name"])["reading_score"].mean().map("{:,.1f}%".format)
twelfth_grader_reading_scores = twelfth_graders.groupby(["school_name"])["reading_score"].mean().map("{:,.1f}%".format)

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.merge(ninth_grader_reading_scores, tenth_grader_reading_scores, on=["school_name"])
reading_scores_by_grade = pd.merge(reading_scores_by_grade, eleventh_grader_reading_scores, on=["school_name"])
reading_scores_by_grade = reading_scores_by_grade.rename(columns={"reading_score_x": "9th Grader Reading Scores", "reading_score_y": "10th Grader Reading Scores", "reading_score": "11th Grader Reading Scores"})
reading_scores_by_grade = pd.merge(reading_scores_by_grade, twelfth_grader_reading_scores, on=["school_name"])
reading_scores_by_grade = reading_scores_by_grade.rename(columns={"reading_score": "12th Grader Reading Scores"})



# Minor data wrangling
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

# Scores by School Spending

```python
# Establish the bins 
bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Budget Per Capita"], bins, labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean(), 1)
spending_reading_scores = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean(),1)
spending_passing_math = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% of students who passed math"].mean(),1)
spending_passing_reading = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% of students who passed reading"].mean(),1)
overall_passing_spending = round(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% of students who passed reading & math"].mean(),1)

```

```python
# Assemble into DataFrame
spending_summary = pd.merge(spending_math_scores, spending_reading_scores, on=["Spending Ranges (Per Student)"])
spending_summary = pd.merge(spending_summary,spending_passing_math, on=["Spending Ranges (Per Student)"])
spending_summary = pd.merge(spending_summary, spending_passing_reading, on=["Spending Ranges (Per Student)"])
spending_summary = pd.merge(spending_summary, overall_passing_spending, on=["Spending Ranges (Per Student)"])

# Display results
spending_summary
```

```python
#Export
spending_summary.to_csv("scores_grouped_by_spending.csv",index=False)
```

# Scores by School Size

```python
# Establish the bins.
bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["# of Students"], bins, labels)
```

```python
per_school_summary.columns
```

```python
# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% of students who passed math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% of students who passed reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% of students who passed reading & math"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.merge(size_math_scores, size_reading_scores, on=["School Size"])
size_summary = pd.merge(size_summary,size_passing_math, on=["School Size"])
size_summary = pd.merge(size_summary, size_passing_reading, on=["School Size"])
size_summary = pd.merge(size_summary, size_overall_passing, on=["School Size"])

# Display results
size_summary
```

```python
#Export
size_summary.to_csv("scores_by_school_size.csv",index=False)
```
